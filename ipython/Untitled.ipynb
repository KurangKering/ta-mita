```python
import numpy as np
import pandas as pd
from random import randrange
from neupy import algorithms
from sklearn.model_selection import train_test_split
```

```python
dataset_norma = pd.read_csv('output/dataset-normalisasi.csv')
dataset = dataset_norma.loc[:, ~dataset_norma.columns.isin(['ID'])]
```

```python
train_df, test_df = train_test_split(dataset, test_size=0.1, stratify=dataset.loc[:, ['CLASS']], random_state=0)
train_df = train_df.sort_index()
test_df = test_df.sort_index()
```
